```python
import xgboost as xgb

booster_params = {'eta':0.05,
                  'gamma': 0.75,
                  'max_depth': 5,
                  'min_child_weight': 3,
                  'max_delta_step': 0,
                  'subsample': 0.85,
                  'colsample_bytree': 1,
                  'tree_method': 'auto',
                  'alpha': 0.25,
                  'lambda': 0.75}

def get_xgb_cv(params, top_k_features = 1):
    X1 = X.loc[:, list(sorted_features[:top_k_features].index)]
    X1.columns = [c.replace('[','(') for c in X1.columns]
    X1.columns = [c.replace(']',')') for c in X1.columns]    
    print(top_k_features)
    dtrain = xgb.DMatrix(data = X1.loc[:,list(sorted_features[:top_k_features].index)].values, label = y.values,nthread= 6)
    xgb_cv = xgb.cv(params = booster_params , dtrain = dtrain, num_boost_round= 300, nfold = 3, stratified = True, metrics = 'auc', seed = 123,verbose_eval=0)
    result = xgb_cv.tail(1).assign(features = list(sorted_features.index)[top_k_features], k = top_k_features)
    return(result)
```

```python
results = pd.concat([get_xgb_cv(booster_params, k) for k in range(1,3500)])
```
