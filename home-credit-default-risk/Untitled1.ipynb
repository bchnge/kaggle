```python
from bayes_opt import BayesianOptimization
```

```python
def fn(x, y):
    return(-x**2 + x/y)
```

```python
bo = BayesianOptimization(fn, {'x' : (-4,3),
                               'y' : (-1,3)})
```

```python
bo.maximize(init_points= 3,n_iter=10)
```

```python
from matplotlib.pyplot import plot
```

```python
plot(bo.X, bo.Y)
```
